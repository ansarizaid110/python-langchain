### prompt templates - dynamic prompts

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
prompts = ChatPromptTemplate.from_messages([
    {"role": "system", "content": "You are a translator and translate input in {language}."},
    {"role": "user", "content": "{query}"}
])

final_prompts = prompts.format_messages(language = "Hinglish", query = "I love to play cricket.")

res = llm.invoke(final_prompts)
print(res.content)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
prompts = ChatPromptTemplate.from_messages([
    {"role": "system", "content": "You are a translator and translate input in {language}."},
    {"role": "user", "content": "{query}"}
])

chains = prompts | llm
```

```python
res = chains.invoke({"language":"Hindi", "query":"I love to play cricket."})
print(res.content)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
out = StrOutputParser()

def transform_case(result:str):
    return result.upper()

prompts = ChatPromptTemplate.from_messages([
    {"role": "system", "content": "You are a translator and translate input in {language}."},
    {"role": "user", "content": "{query}"}
])

chains = prompts | llm | out | transform_case
res = chains.invoke({"language":"Hinglish", "query":"I love to play cricket."})
print(res)
```
