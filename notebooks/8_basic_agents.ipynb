```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a: int, b: int):
    """
        It will return sum of two numbers.
        Args:
            a: Number One
            b: Number Two
    """
    return a + b

@tool
def multiply_numbers(a: int, b: int):
    """
        It will return product of two numbers.
        Args:
            a: Number One
            b: Number Two
    """
    return a * b
```

```python
add_numbers.invoke({"a": 10, "b": 20})
```

```python
from langchain_openai import ChatOpenAI
from langchain_groq import ChatGroq
from langchain.agents import create_agent
import re
```

```python
llm = ChatOpenAI(model="gpt-4o")
llm_groq = ChatGroq(model="openai/gpt-oss-120b")
```

```python
agent = create_agent(
    model=llm_groq,
    tools=[add_numbers, multiply_numbers],
    system_prompt="You are a math teacher and always use tool for calculation."
)
```

```python
response = agent.invoke({"messages": [{"role": "user", "content": "What is 2 + 30 * 10 ?"}]})
print(re.sub(r'[^a-zA-Z0-9 ]', '', response["messages"][-1].content))
```
