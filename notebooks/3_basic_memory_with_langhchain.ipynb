```python
from dotenv import load_dotenv
load_dotenv()

from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o")

prompts = [
    {"role": "user", "content": "Hello, My name is Zaid Ahmed."},
    {"role": "ai", "content": "Hello Zaid! How can I assist you today."},
    {"role": "user", "content": "What is my name ?"}
]

res = llm.invoke(prompts)

res.content
```

### Manage History

```python
history = []
while True:
    query = input("User: ")
    if query.lower() in ["exit", "quit", "bye"]:
        print("Good Bye 👋")
        break

    history.append({"role": "user", "content": query})
    print("User: ", query)

    res = llm.invoke(history)
    history.append({"role": "ai", "content": res.content})
    print("AI: ", res.content, "\n")
```
